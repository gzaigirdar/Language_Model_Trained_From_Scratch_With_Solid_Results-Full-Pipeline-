```python
import matplotlib.pyplot as plt 
import numpy as np
```

```python
inputs = [d['input'] for d in dialogues]
outputs = [d['output'] for d in dialogues]

rows = list(zip(inputs, outputs))

fig, ax = plt.subplots(figsize=(10, len(rows)*0.5))
ax.axis('off')

table = ax.table(
    cellText=rows,
    colLabels=['Input', 'Output'],
    cellLoc='left',
    loc='center'
)

table.auto_set_font_size(False)
table.set_fontsize(10)
table.scale(1, 1.5)

plt.show()
```
